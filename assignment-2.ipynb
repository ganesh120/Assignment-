```python
#Q2 Write a program to accept the cost price of a bike and display
#the road tax to be paid according to the following criteria
```

```python
# Function to calculate tax based on cost price
def calculate_tax(cost_price):
    if cost_price > 100000:
        tax = 0.15 * cost_price  # 15% tax
    elif 50000 < cost_price <= 100000:
        tax = 0.10 * cost_price  # 10% tax
    elif cost_price == 50000:
        tax = 0.05 * cost_price  # 5% tax
    else:
        tax = 0  # No tax for cost price below 50000
    return tax

# For example, we will put directly  value to cost_price
cost_price = 55000  # Replace with any value for testing

# Calculate the tax to be paid
tax_to_be_paid = calculate_tax(cost_price)

# Display the tax to be paid
print(f"Cost price of the bike: Rs {cost_price}")
print(f"Tax to be paid: Rs {tax_to_be_paid:.2f}")

```

```python
#q4 check how many times a given number can be divided by 3 before it is less than or equal to 10

number = 100  

# how many count
count = 0

# Loop until the number is less than or equal to 10
while number > 10:
    number /= 3  # Divide the number by 3
    count += 1   # Increment count

# Print the result
print("The number can be divided by 3", count, "times before it is less than or equal to 10.")

```

```python
#Q6 Use nested while loop to print 3 different pattern

#pattern 1 
print("Pattern 1")
#initialize variables 
i=1
j=1
#while loop
while i<5:
    #nested while
    while j<=i:
        #printed j without space
        print(j, end="")
        #increment j by 1
        j+=1
    #print new line
    print()
    #reset value of j
    j=1
    #increment value of i by 1
    i+=1

print("----------------------------------------------------------------")
print("Pattern 2")
#initialize variables 
i=1
j=1
#while loop
while i<5:
    #nested while
    while j<=i:
        #printed * without space
        print("*", end="")
        #increment j by 1
        j+=1
    #print new line
    print()
    #reset value of j
    j=1
    #increment value of i by 1
    i+=1

print("----------------------------------------------------------------")
print("Pattern 3")
#initialize variables 
i=1
j=1
#while loop
while i<10:
    #nested while
    while j<10:
        #printed smile with smile using Unicode associated with it
        print("\U0001f600\U0001f600", end="\U0001F606")
        #increment j by 1
        j+=1
    #print new line
    print()
    #reset value of j
    j=1
    #increment value of i by 1
    i+=1
```
